```python
# Remove commas and currency symbols, if any
customer['Income'] = customer['Income'].replace('[\$,]', '', regex=True)
```

```python
customer['Income'] = pd.to_numeric(customer['Income'], errors='coerce')
```

```python
customer['Income'].fillna(customer['Income'].median(), inplace=True)
```
